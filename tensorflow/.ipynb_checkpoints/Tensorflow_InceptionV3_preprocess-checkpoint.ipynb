# Tensorflow_InceptionV3_preprocess
在本脚本中，我们测试了两种预处理方法对InceptionV3的影响
1. Slim函数自带的Inception预处理方法
2. 论文方法，将输入图像归一化成【-1,1】

## 结论：
在本测试图片中，我发现第二种方法的预处理要优于Slim自带的预处理方法

================================================================================================ 

### 方法一：Slim函数自带的Inception预处理算法：
- 将像素值转换到[0, 1]区间内。
- 中心切片。
- resize图片。
- 将像素值转换到[-1, 1]区间内。

```python
import tensorflow as tf
from tensorflow.contrib.slim import nets
slim = tf.contrib.slim
from PIL import Image 
import numpy as np
from preprocessing import inception_preprocessing
```

```python
inputs = tf.placeholder("float32", shape=[None, None, 3], name="inputs")
preprocess_inputs = inception_preprocessing.preprocess_image(inputs, 299, 299, is_training=False)
preprocess_inputs = tf.expand_dims(preprocess_inputs, 0)
with slim.arg_scope(nets.inception.inception_v3_arg_scope()):
    net, endpoints = nets.inception.inception_v3(preprocess_inputs, num_classes=1001, is_training=False)
prob = tf.nn.softmax(net)
```

```python
sess = tf.InteractiveSession()
# saver = tf.train.Saver()
# saver.restore(sess, "inception_v3.ckpt")
variables_to_restore = slim.get_variables_to_restore()
init_fn = slim.assign_from_checkpoint_fn("inception_v3.ckpt", variables_to_restore, ignore_missing_vars=True)
init_fn(sess)
```

```python
im = Image.open("1.jpg")
im = np.array(im) / 255.
```

```python
# show the image feed the network
preprocess_im = sess.run(preprocess_inputs, feed_dict={inputs:im})
Image.fromarray(np.uint8((preprocess_im[0]/2+0.5)*255.0))
```

```python
t = sess.run(endpoints["Predictions"], feed_dict={inputs: im})
print("find the top1 prediction",np.where(t==np.max(t)))
print("print prob: ",t[0][387])
```

### 方法二：直接将输入图像归一化成【-1,1】

```python
import tensorflow as tf
from tensorflow.contrib.slim import nets
slim = tf.contrib.slim
from PIL import Image 
import numpy as np
from preprocessing import inception_preprocessing
```

```python
inputs = tf.placeholder("float32", shape=[None, 299, 299, 3], name="inputs")
# preprocess_inputs = inception_preprocessing.preprocess_image(inputs, 299, 299, is_training=False)
# preprocess_inputs = tf.expand_dims(preprocess_inputs, 0)
with slim.arg_scope(nets.inception.inception_v3_arg_scope()):
    net, endpoints = nets.inception.inception_v3(inputs, num_classes=1001, is_training=False)
prob = tf.nn.softmax(net)
```

```python
sess = tf.InteractiveSession()
# saver = tf.train.Saver()
# saver.restore(sess, "inception_v3.ckpt")
variables_to_restore = slim.get_variables_to_restore()
init_fn = slim.assign_from_checkpoint_fn("inception_v3.ckpt", variables_to_restore, ignore_missing_vars=True)
init_fn(sess)
```

```python
# show the image feed the network
Image.open("1.jpg").resize((299,299),Image.NEAREST)
```

```python
im = Image.open("1.jpg").resize((299,299),Image.NEAREST)
im = np.array(im, dtype=np.float32) / 127.5 - 1
im = np.expand_dims(im, axis=0)
```

```python
t = sess.run(endpoints["Predictions"], feed_dict={inputs:im})
```

```python
print("find the top1 prediction",np.where(t==np.max(t)))
print("print prob: ",t[0][387])
```
